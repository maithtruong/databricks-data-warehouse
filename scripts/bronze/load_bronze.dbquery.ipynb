```sql
/*
  Stored Procedure: Load Bronze
  Description:
    Create a procedure that load raw data files (.csv format) into tables in the bronze schema.
    Input: the names of source files in the source volumes.
    Method: Bulk Insert & Full Load.
  WARNING:
    Running this procedure will drop all rows in the tables in the bronze schema and force a rewrite using the version of the source files at the time of the procedure call.
    *This procedure modifies the default behavior of COPY INTO: It is by default idempotent, which means that if you run the same procedure on the same file names, the tables will not be updated.
    With this modification users can force a rewrite of the tables using the files at the same path, with the same name.
*/

CREATE OR REPLACE PROCEDURE load_bronze()
  LANGUAGE SQL
  SQL SECURITY INVOKER AS
  BEGIN

    SELECT '=================================================';
    SELECT 'Loading bronze layer...';
    SELECT '=================================================';

    SELECT '-------------------------------------------------';
    SELECT 'Loading CRM tables...';
    SELECT '-------------------------------------------------';

    -- CRM Tables
    SELECT '>> Truncating and adding new rows to: bronze.crm_cust_info...';
    TRUNCATE TABLE core_dev.bronze.crm_cust_info;
    COPY INTO core_dev.bronze.crm_cust_info
    FROM '/Volumes/core_dev/source/source_crm/cust_info.csv'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true', 'nullValue' = '')
    COPY_OPTIONS ('force' = 'true');

    SELECT '>> Truncating and adding new rows to: bronze.crm_prd_info...';
    TRUNCATE TABLE core_dev.bronze.crm_prd_info;
    COPY INTO core_dev.bronze.crm_prd_info
    FROM '/Volumes/core_dev/source/source_crm/prd_info.csv'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true', 'nullValue' = '')
    COPY_OPTIONS ('force' = 'true');

    SELECT '>> Truncating and adding new rows to: bronze.crm_sales_details...';
    TRUNCATE TABLE core_dev.bronze.crm_sales_details;
    COPY INTO core_dev.bronze.crm_sales_details
    FROM '/Volumes/core_dev/source/source_crm/sales_details.csv'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true', 'nullValue' = '')
    COPY_OPTIONS ('force' = 'true');

    SELECT '-------------------------------------------------';
    SELECT 'Loading ERP tables...';
    SELECT '-------------------------------------------------';

    -- ERP Tables
    SELECT '>> Truncating and adding new rows to: bronze.erp_cust_az12...';
    TRUNCATE TABLE core_dev.bronze.erp_cust_az12;
    COPY INTO core_dev.bronze.erp_cust_az12
    FROM '/Volumes/core_dev/source/source_erp/CUST_AZ12.csv'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true', 'nullValue' = '')
    COPY_OPTIONS ('force' = 'true');

    SELECT '>> Truncating and adding new rows to: bronze.erp_loc_a101...';
    TRUNCATE TABLE core_dev.bronze.erp_loc_a101;
    COPY INTO core_dev.bronze.erp_loc_a101
    FROM '/Volumes/core_dev/source/source_erp/LOC_A101.csv'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true', 'nullValue' = '')
    COPY_OPTIONS ('force' = 'true');

    SELECT '>> Truncating and adding new rows to: bronze.erp_px_cat_g1v2...';
    TRUNCATE TABLE core_dev.bronze.erp_px_cat_g1v2;
    COPY INTO core_dev.bronze.erp_px_cat_g1v2
    FROM '/Volumes/core_dev/source/source_erp/PX_CAT_G1V2.csv'
    FILEFORMAT = CSV
    FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true', 'nullValue' = '')
    COPY_OPTIONS ('force' = 'true');

    SELECT 'SUCCESS. All tasks performed successfully.';
END;

-- Example
-- CALL load_bronze();
```

```sql
/*
  DDL Script: Create Bronze Tables
  Description:
    This script creates the tables inside the bronze schema, using a format similar to the source data.
    It leverages Databricks's Delta format for efficient storage and processing.
*/

-- Choose catalog and schema
USE CATALOG core_dev;
USE SCHEMA bronze;

-- crm_cust_info
CREATE OR REPLACE TABLE bronze.crm_cust_info (
    cst_id INT,
    cst_key STRING,
    cst_firstname STRING,
    cst_lastname STRING,
    cst_marital_status STRING,
    cst_gndr STRING,
    cst_create_date DATE
)
USING DELTA;

-- crm_prd_info
CREATE OR REPLACE TABLE bronze.crm_prd_info (
    prd_id INT,
    prd_key STRING,
    prd_nm STRING,
    prd_cost INT,
    prd_line STRING,
    prd_start_dt DATE,
    prd_end_dt DATE
)
USING DELTA;

-- crm_sales_details
CREATE OR REPLACE TABLE bronze.crm_sales_details (
    sls_ord_num STRING,
    sls_prd_key STRING,
    sls_cust_id INT,
    sls_order_dt INT,
    sls_ship_dt INT,
    sls_due_dt INT,
    sls_sales INT,
    sls_quantity INT,
    sls_price INT
)
USING DELTA;

-- erp_loc_a101
CREATE OR REPLACE TABLE bronze.erp_loc_a101 (
    cid STRING,
    cntry STRING
)
USING DELTA;

-- erp_cust_az12
CREATE OR REPLACE TABLE bronze.erp_cust_az12 (
    cid STRING,
    bdate DATE,
    gen STRING
)
USING DELTA;

-- erp_px_cat_g1v2
CREATE OR REPLACE TABLE bronze.erp_px_cat_g1v2 (
    id STRING,
    cat STRING,
    subcat STRING,
    maintenance STRING
)
USING DELTA;
```
